```python
import pandas as pd

# https://www.kaggle.com/datasets/oscarm524/predicting-returns-of-discounted-articles-sales?resource=download
df = pd.read_csv("./raw/orders_train.txt", sep=";")

print(df.dtypes)
df
```

```python
from sqlalchemy import create_engine, text

engine = create_engine("postgresql+psycopg2://aaa:111@localhost:5433/ml_course_project")

df.to_sql("raw_orders", engine, if_exists="replace", index=False)
```

```python
pd.read_sql("select * from raw_orders", engine)
```

```python
pd.read_sql("""
select
    count(*) as rows,
    count(distinct "customerID") as users,
    count(distinct "articleID") as items
from raw_orders
""", engine)
```
